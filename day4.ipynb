```python
plt.style.use('seaborn-v0_8-darkgrid')
plt.figure(figsize=(10, 6))
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y, 'g-', label='$y1=sin(x)$')
plt.plot(x,z, 'bo', label='$y2=cos(x)$')
plt.xlabel('x')
plt.ylabel('Function value')
plt.title('sin(x) and cos(x)',fontsize=14)
plt.legend(loc='upper right',bbox_to_anchor=(1.2,1.0))
plt.show()
```

Multiple plots 

```python
plt.style.use('seaborn-v0_8-darkgrid')
plt.figure(figsize=(12,5))
x = np.linspace(0, 10, 100) 
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x, y1,'g-', label='sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine wave')
plt.legend(loc='upper right')

plt.subplot(2,1,2)
plt.plot(x, y2,'ro', label='cos(x)')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.title('Cosine wave')
plt.legend(loc='upper right')
```

```python
x = np.arange(-50,50)
y1 = np.power(x,2)
y2 = np.power(x,3)
y3 = np.sin(x)
y4 = np.cos(x)
y5 = np.tan(x)
y6 = np.tanh(x)
y7 = np.sinh(x)
y8 = np.cosh(x)
y9 = np.exp(x)

fig, ax1 = plt.subplots(nrows=3, ncols=3, figsize=(20,10))
ax1[0,0].plot(x, y1, 'tab:orange')
ax1[0,1].plot(x, y2, 'tab:green')
ax1[0,2].plot(x, y3, 'b-')
ax1[1,0].plot(x, y4, 'r-')
ax1[1,1].plot(x, y5, 'g-')
ax1[1,2].plot(x, y6, 'm-')
ax1[2,0].plot(x, y7, 'y-')
ax1[2,1].plot(x, y8, 'k-')
ax1[2,2].plot(x, y9, 'k-')
```

Usage of loops is also possible

```python
x = [4,5,6,7,8]
y = [[1,2,3,4,5],[10,20,30,40,50],[1,4,9,16,25]]
plt.figure(figsize=(10,6))
for i in y:
      plt.plot(x,i)
```

```python
id1=np.arange(1,10)
score=np.arange(20,110,10)
plt.bar(id1,score)
plt.xlabel('id')
plt.ylabel('score')
```

```python
id1 = np.arange(1,10)
score = np.arange(20,110,10)
plt.figure(figsize=(8,5))
ax = plt.axes()
ax.set_facecolor("black")
plt.bar(id1,score,color='red',width=0.5)
plt.title('Score of Students')
plt.xlabel('Student ID')
plt.ylabel('Score')
plt.show()
```
