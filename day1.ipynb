````python
x=[1,2,3,4,5,6,7,8]
y=[10,8,7,9,4,3,7,8]
plt.title("Sample plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x,y,marker="x",color='red')
plt.plot(x,y,linestyle="-.",color='black')
plt.legend(['data points'],loc='upper right')
plt.savefig('sampleplot.png')
plt.show()
````

This plot uses different colors for each point and a diamond marker to clearly distinguish values. Good for categorical representation.

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [5, 8, 2, 6, 9]
colors = ['red', 'blue', 'green', 'orange', 'purple']

plt.scatter(x, y, c=colors, s=100, marker='D')  # Diamond markers with colors
plt.title('Scatter Plot with Colored Markers')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
````

 Bar chart with custom hex colors. Great for categorical data visualization.

````python
x = ['A', 'B', 'C', 'D']
y = [3, 7, 5, 9]
colors = ['#FF5733', '#33FFBD', '#FF33EC', '#3385FF']

plt.bar(x, y, color=colors, width=0.5)
plt.title('Custom Colored Bar Chart')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
````

Adds a text label with an arrow pointing to a specific data point. Helpful for emphasizing key information.

````python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, marker='o', linestyle='-', color='green')
plt.title('Line Plot with Annotations')
plt.xlabel('X')
plt.ylabel('Y')

# Annotate a point
plt.annotate('Mid-point', xy=(3, 6), xytext=(3.5, 7),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.grid(True)
plt.show()
````

Compares two mathematical functions with different styles and a legend to distinguish them.

````python
x = [0, 1, 2, 3, 4, 5]
y1 = [0, 1, 4, 9, 16, 25]
y2 = [0, 2, 8, 18, 32, 50]

plt.plot(x, y1, label='y = xÂ²', color='blue', marker='s')
plt.plot(x, y2, label='y = 2xÂ²', color='red', linestyle='--', marker='^')

plt.title('Comparing Two Functions')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()
plt.grid(True)
plt.show()
````
