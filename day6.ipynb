# Ticks, Stylesheets: A Data Visualization Tutorial

## 1. Importing Libraries

```python
# Explanation: This cell demonstrates a concept related to ticks or stylesheets.
import matplotlib.pyplot as plt
import numpy as np
```

## 2. Creating Sample Data


```python
# Explanation: This cell demonstrates a concept related to ticks or stylesheets.
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

## 3. Plotting Data with Custom Ticks and Stylesheets

```python
# Explanation: This cell demonstrates a concept related to ticks or stylesheets.
plt.figure(figsize=(10, 5))
plt.plot(x, y1, label='Sine Wave', color='blue')
plt.plot(x, y2, label='Cosine Wave', color='orange')
plt.title('Sine and Cosine Waves')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xticks(np.arange(0, 11, 1))  # Custom x-ticks
plt.yticks(np.arange(-1, 2, 0.5))  # Custom y-ticks
plt.grid(True)
plt.legend()
plt.xlim(0, 10)
plt.ylim(-1, 1)
plt.show()
```

## 4. Applying Stylesheets

```python
# Explanation: This cell demonstrates a concept related to ticks or stylesheets.
plt.style.use('ggplot')  # Applying ggplot style
plt.figure(figsize=(10, 5))
plt.plot(x, y1, label='Sine Wave', color='blue')
plt.plot(x, y2, label='Cosine Wave', color='orange')
plt.title('Sine and Cosine Waves with ggplot Style')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xticks(np.arange(0, 11, 1))  # Custom x-ticks
plt.yticks(np.arange(-1, 2, 0.5))  # Custom y-ticks
plt.grid(True)
plt.legend()
plt.xlim(0, 10)
plt.ylim(-1, 1)
plt.show()
```

## 5. Adding Annotations

```python
# Explanation: This cell demonstrates a concept related to ticks or stylesheets.
plt.figure(figsize=(10, 5))
plt.plot(x, y1, label='Sine Wave', color='blue')
plt.plot(x, y2, label='Cosine Wave', color='orange')
plt.title('Sine and Cosine Waves with Annotations')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xticks(np.arange(0, 11, 1))  # Custom x-ticks
plt.yticks(np.arange(-1, 2, 0.5))  # Custom y-ticks
plt.grid(True)
plt.xlim(0, 10)
plt.ylim(-1, 1)
plt.legend()
plt.annotate('Peak of Sine Wave', xy=(np.pi/2, 1), xytext=(np.pi/2 + 1, 1.2),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('Peak of Cosine Wave', xy=(0, 1), xytext=(1, 1.2),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

## Additional Example: Using 'seaborn-poster' Style with Customized Ticks

```python

import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Applying a different stylesheet
plt.style.use('seaborn-poster')

fig, ax = plt.subplots()
ax.plot(x, y, color='green', marker='o', label='sin(x)')

# Customizing ticks: placing them on top and right as well
ax.tick_params(axis='both', which='both', direction='inout', length=10, width=2, colors='red', labelsize=12, top=True, right=True)

# Adding grid for better visibility
ax.grid(True, linestyle='--', alpha=0.7)

ax.set_title('Customized Ticks with seaborn-poster Style')
ax.set_xlabel('X-axis')
ax.set_ylabel('sin(x)')
ax.legend()
plt.show()
```
